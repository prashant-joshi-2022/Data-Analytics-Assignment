## **Ques 1)-** Explain the difference between Relative , Absolute and Mixed Cell Referencing.

### **Answer :**
* Relative cell referencing (A1) - changes both the row and column when a formula is copied to another cell, making it useful when the same formula needs to adjust automatically for different rows or columns.

* Absolute cell referencing ($A$1) - does not change at all when copied because both the row and column are fixed using dollar signs, so it is used when a constant value is required in every formula.

* Mixed cell referencing ($A1 or A$1) - changes only one part of the reference while keeping the other fixed, which helps when working with tables or calculation grids where either the row or the column should remain constant.
